```python
from pandas import DataFrame

from lib.evaluation.time.time_statistics import get_time_stats
from lib.utils.export_table import export_table_to_csv, export_table_to_latex
from lib.utils.thesis_names import get_is_parser_api

table_name = "time_per_page"

df = get_time_stats()
```

```python
df = df.rename(columns={"mean": r"$\mu$", "std": r"$\sigma$"}, level=1)
```

```python
export_table_to_latex(df, table_name, highlight_mode=False, column_format="lcc", escape_latex=False)
```

```python
parsing_mean: DataFrame = df[df.columns[0]]
parsing_mean.rename(index=get_is_parser_api, inplace=True)
parsing_mean = parsing_mean.reset_index()
parsing_mean.columns = ["Method", "Mean"]

parsing_mean.sort_values(
    by=["Method", "Mean"],
    key=lambda col: col.str.endswith("*") if col.name == "Method" else col,
    inplace=True
)

parsing_mean
```

```python
export_table_to_csv(parsing_mean, table_name)
```
